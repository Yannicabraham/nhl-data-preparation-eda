## Part 1 - Data Coding and Merging

```python
#Import Libraries
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import NHL Team Data and Display
NHL_Team=pd.read_csv("Assignment Data/Week 2/NHL_team.csv")
display(NHL_Team)
```

```python
#Deleting unnecessary rows
NHL_Team.drop(columns=['Unnamed: 0','abbr', 'tname','lname','sname'], axis=1, inplace=True)
display(NHL_Team)
```

```python
#Renaming for clarity
NHL_Team.rename(columns={'name':'team_name'}, inplace=True)
display(NHL_Team)
```

```python
#Import NHL Competition Data and Display Head
NHL_Competition=pd.read_csv("Assignment Data/Week 2/NHL_competition.csv")
NHL_Competition.head()
```

```python
#Deleting unwanted columns
NHL_Competition.drop(columns=['Unnamed: 0','tz','start','end'], axis=1,inplace=True)
display(NHL_Competition)
```

```python
#Renaming the variable ‚Äúname‚Äù to ‚Äúcompetition_name‚Äù for clarity again
NHL_Competition.rename(columns={'name':'competition_name'}, inplace=True)
display(NHL_Competition)
```

```python
#Import NHL Game Data and Display Head
NHL_Game=pd.read_csv("Assignment Data/Week 2/NHL_game.csv")
NHL_Game.head()
```

```python
#Deleting the following variables: ‚ÄúX‚Äù, ‚Äúperiod‚Äù, and ‚Äústatus‚Äù.
NHL_Game.drop(columns=['X','period','status'], axis=1, inplace=True)
display(NHL_Game)
```

```python
#Merging the ‚ÄúNHL_Team‚Äù df into the ‚ÄúNHL_Game‚Äù df by ‚Äútid.‚Äù Equals updated "NHL_Game" df.
NHL_Game = pd.merge(NHL_Team,NHL_Game, on=['tid'])
display(NHL_Game)
```

```python
#Merging df ‚ÄúNHL_Competition‚Äù into the df ‚ÄúNHL_Game‚Äù by ‚Äúcomp_id.‚Äù Equals updated "NHL_Game" df.
NHL_Game = pd.merge(NHL_Competition, NHL_Game, on=['comp_id'])
display(NHL_Game)
```

```python
#In the merged ‚ÄúNHL_Game‚Äù df, we create the new variable ‚Äúhgd‚Äù to indicate the goal difference between home and away 
#score (hscore ‚Äì ascore). We then delete observations with missing values in the variable ‚Äúhgd‚Äù.
NHL_Game['hgd'] = NHL_Game['hscore']-NHL_Game['ascore']
NHL_Game=NHL_Game[pd.notna(NHL_Game['hgd'])]
display(NHL_Game)
```

```python
#Checking for remaining 0 values in the "NHL_Game" df
NHL_Game.isnull().any()
```

```python
NHL_Game.shape
```

```python
#Converting the type of the ‚Äúdate‚Äù variable from ‚Äúobject‚Äù to ‚Äúdatetime.‚Äù
NHL_Game.dtypes
```

```python
NHL_Game['date']=pd.to_datetime(NHL_Game['date'])
NHL_Game.head()
```

```python
#Sorting the NHL games by ‚Äúdate‚Äù and limiting to the first 15 observations.
NHL_Game.sort_values(by='date').head(15)
```

```python
#Create√≠ng two dfs that separate the ‚ÄúNHL_Game‚Äù dataframe by home and away games. 

NHL_Home = NHL_Game[NHL_Game['home_away'] == 'home']
NHL_Away = NHL_Game[NHL_Game['home_away']== 'away']
```

```python
#Renaming variables: 
#For away games: ‚Äúascore‚Äù to ‚Äúgoals_for‚Äù; ‚Äúhscore‚Äù to ‚Äúgoals_against‚Äù
#For home games: ‚Äúhscore‚Äù to ‚Äúgoals_for‚Äù; ‚Äúascore‚Äù to ‚Äúgoals_against‚Äù 

NHL_Away.rename(columns={'ascore':'goals_for','hscore':'goals_against'},inplace=True)
NHL_Home.rename(columns={'hscore':'goals_for','ascore':'goals_against'},inplace=True)
display(NHL_Away)
```

```python
#Creating a new ‚Äúwin‚Äù variable that equals to 1 if the team won the game; 0 if the team lost the game; 
#and 0.5 if it was a draw. 

NHL_Away['win'] = np.where(NHL_Away['goals_for'] > NHL_Away['goals_against'], 1, np.where(NHL_Away['hgd'] == 0,0.5,0))
display(NHL_Away)
```

```python
NHL_Home['win'] = np.where(NHL_Home['goals_for'] > NHL_Home['goals_against'], 1, np.where(NHL_Home['hgd'] == 0,0.5,0))
display(NHL_Home)
```

```python
#Appending the ‚ÄúNHL_Home‚Äù and ‚ÄúNHL_Away‚Äù dataframes to be the new ‚ÄúNHL_Game‚Äù dataframe using the concat function
NHL_Game = pd.concat([NHL_Home, NHL_Away], axis=0, ignore_index=True)
display(NHL_Game)
```

```python
#Generating a team level dataframe that aggregates the total number of games won, the total number of 
#‚Äúgoals_for‚Äù and ‚Äúgoals_against‚Äù for each team in each competition (i.e. grouped by tid, competition_name and type)
#Naming this new dataframe ‚ÄúNHL_Team_Stats‚Äù. 
#Making sure to convert the indexes of the new dataframe back as  variables. 

NHL_Team_Stats = NHL_Game.groupby(['team_name','tid','competition_name','type'])[['win','goals_for','goals_against']].sum().sort_values(['tid','competition_name']).reset_index()
display(NHL_Team_Stats)
```

```python
#Creating a dataframe ‚ÄúNHL_Game_Count‚Äù that include the total number of games played by each team in each 
#competition (i.e. grouped by tid, competition_name and type). 
#Naming this new variable in the dataframe ‚Äúgame_count‚Äù

NHL_Game['game_count'] = 1
NHL_Game_Count = NHL_Game.groupby(['team_name','tid','competition_name','type'])[['game_count']].sum().sort_values(['tid','competition_name']).reset_index()
display(NHL_Game_Count)
```

```python
#Merging dataframes.
#(i) Merging the ‚ÄúNHL_Game_Count‚Äù df into the ‚ÄúNHL_Team_Stats‚Äù df by ‚Äútid‚Äù, ‚Äúcompetition_name‚Äù, and ‚Äútype‚Äù. 
#Continue to name the merged df ‚ÄúNHL_Team_Stats‚Äù.

NHL_Team_Stats = pd.merge(NHL_Game_Count, NHL_Team_Stats, on=['tid','competition_name','type','team_name'])
display(NHL_Team_Stats)
```

```python
#(ii) Merging the ‚ÄúNHL_Team‚Äù df into the ‚ÄúNHL_Team_Stats‚Äù df by ‚Äútid‚Äù. 
#Continue to name the merged dataframe ‚ÄúNHL_Team_Stats‚Äù.

NHL_Team_Stats = pd.merge(NHL_Team,NHL_Team_Stats, on=['tid', 'team_name'])
display(NHL_Team_Stats)
```

```python
#Import NHL PPPK Data and Display Head
NHL_PPPK=pd.read_csv("Assignment Data/Week 2/pp.pk.ppgf.csv")
NHL_PPPK.head()
```

```python
#Merging the ‚ÄúNHL_PPPK‚Äù dataframe into the ‚ÄúNHL_Team_Stats‚Äù dataframe by ‚Äútricode‚Äù and ‚Äúcompetition_name‚Äù.
NHL_Team_Stats = pd.merge(NHL_Team_Stats, NHL_PPPK, on=['tricode','competition_name'])
display(NHL_Team_Stats)
```

```python
#Creating new variables in the ‚ÄúNHL_Team_Stats‚Äù dataframe.
#(i)Winning percentage (‚Äúwin_pct‚Äù)=‚Äùwin‚Äù/ total number of games played
#(ii)Average goals for per game (‚Äúavg_gf‚Äù)=total number of goals for / total number of games played     
#(iii)Average goals against per game (‚Äúavg_ga‚Äù)=total number of goals against / total number of games played

NHL_Team_Stats['win_pct']=NHL_Team_Stats['win']/NHL_Team_Stats['game_count']
NHL_Team_Stats['avg_gf']=NHL_Team_Stats['goals_for']/NHL_Team_Stats['game_count']
NHL_Team_Stats['avg_ga']=NHL_Team_Stats['goals_against']/NHL_Team_Stats['game_count']
display(NHL_Team_Stats)
```

```python
#In the ‚ÄúNHL_Competition‚Äù dataframe, the variable ‚Äútype‚Äù indicates the type of competition:
#type=2 ‚Äì regular season. Create a dataframe that contains team statistics for games only during regular seasons.
#Name this dataframe ‚ÄúNHL_Team_R_Stats‚Äù. 

NHL_Team_R_Stats = NHL_Team_Stats[NHL_Team_Stats['type']==2]
display(NHL_Team_R_Stats)
```

## Part 2 - Descriptive and Summary Analyses

```python
#In the ‚ÄúNHL_Game‚Äù dataframe, calculate summary statistics for the  ‚Äúgoals_for‚Äù variable; 
NHL_Game['goals_for'].describe()
```

```python
#calculate summary statistics for the ‚Äúgoals_against‚Äù variable based on whether it is home or away game.
NHL_Game.groupby(['home_away'])['goals_against'].describe()
```

```python
#Create a histogram of the ‚Äúgoals_against‚Äù variable by whether the game is home or away 
#(i)Make the color of the histogram green
#(ii)Set the number of bins to be 20
#(iii)Make sure the two sub-histograms share the same ranges for the x-axis and y-axis.

NHL_Game.hist(by='home_away',column='goals_against',color='green',bins=20, sharex=True,sharey=True)
plt.show()
```

## Part 3 - Correlation Analyses


```python
#In the ‚ÄúNHL_Team_R_Stats‚Äù dataframe, make a scatter plot to depict the relationship between the total number of goals for and the winning percentage. 
#(i)Plot the total number of goals for on the x-axis and winning percentage on the y-axis.
#(ii)Add a regression line to the scatter plot.
#(iii)Make the title of the graph ‚ÄúRelationship between Goals for and Winning Percentage‚Äù and make the font size 11.
#(iv)Label the x-axis ‚ÄúTotal Goals for‚Äù and label the y-axis ‚ÄúWinning Percentage‚Äù.

sns.regplot(x='goals_for',y='win_pct', data=NHL_Team_R_Stats, marker='.')
plt.xlabel('Total Goals for')
plt.ylabel('Winning %')
plt.title('Relationship Between Goals for and Winning %', fontsize=11)
plt.show()
```

```python
#In the ‚ÄúNHL_Team_R_Stats‚Äù dataframe, calculate the correlation coefficient between total number of 
#goals for and winning percentage.

NHL_Team_R_Stats['goals_for'].corr(NHL_Team_R_Stats['win_pct'])
```

```python
#Create a scatter plot of the total number of goals for and winning percentage similar to step 1 for 
#regular season games. In this graph, group observations by ‚Äúcompetition_name‚Äù.
#(i)Plot the total number of goals for on the x-axis and winning percentage on the y-axis.
#(ii)Add a regression line to the scatter plot.
#(iii)Make the title of the graph ‚ÄúRelationship between Goals for and Winning Percentage‚Äù and make the font size 11.
#(iv)Label the x-axis ‚ÄúTotal Goals for‚Äù and label the y-axis ‚ÄúWinning Percentage‚Äù.

sns.lmplot(x='goals_for',y='win_pct', hue='competition_name', data=NHL_Team_Stats)
plt.title('Relationship between Goals for and Winning Percentage', fontsize=11)
plt.xlabel('Total Goals for')
plt.ylabel('Winning Percentage')
plt.show()
```

```python
#For the ‚ÄúNHL_Team_R_Stats‚Äù dataframe, delete observations of 2011 and 2012 seasons. 
#Continue to name the dataframe ‚ÄúNHL_Team_R_Stats‚Äù.

NHL_Team_R_Stats = NHL_Team_R_Stats[
    (NHL_Team_R_Stats.competition_name != '2011 NHL Regular Season') &
    (NHL_Team_R_Stats.competition_name != '2011 NHL Playoff') &
    (NHL_Team_R_Stats.competition_name != '2012 NHL Regular Season') &
    (NHL_Team_R_Stats.competition_name != '2012 NHL Playoff')]
display(NHL_Team_R_Stats)
```

```python
#In the new ‚ÄúNHL_Team_R_Stats‚Äù dataframe, create a scatter plot of total number of goals for and 
#winning percentage.
#(i)Plot the total number of goals for on the x-axis and winning percentage on the y-axis.
#(ii)Add a regression line to the scatter plot.
#(iii)Make the title of the graph ‚ÄúRelationship between Goals for and Winning Percentage‚Äù and make the font size 11.
#(iv)Label the x-axis ‚ÄúTotal Goals for‚Äù and label the y-axis ‚ÄúWinning Percentage‚Äù.

sns.regplot(x='goals_for',y='win_pct',data=NHL_Team_R_Stats)
plt.title('Relationship Between Goals For and Winning Percentage', fontsize=11)
plt.ylabel('Winning Percentage')
plt.xlabel('Total Goals For')
plt.show()
```

```python
#Calculate the correlation coefficient between total number of goals for and winning percentage 
#in the updated ‚ÄúNHL_Team_R_Stats‚Äù dataframe.

NHL_Team_R_Stats['goals_for'].corr(NHL_Team_R_Stats['win_pct'])
```
